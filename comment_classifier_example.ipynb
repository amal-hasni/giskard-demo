```python
import comments_classifier_example
from giskard import Model, Dataset, scan
from sklearn.metrics import classification_report


# Get Test Dataset and Model
df_train, df_test = comments_classifier_example.comments_datasets()
clf_func, classes = comments_classifier_example.get_trained_model(df_train)
```

```python
df_test.head()
```

```python
# Create Giskard Dataset and Model
giskard_dataset = Dataset(
    df=df_test,
    target='target',
    name="Comments dataset",
)

giskard_model = Model(
    model=clf_func,
    model_type="classification",
    name="Toxicity model",
    classification_labels=classes,
)


results = scan(giskard_model, giskard_dataset, verbose=False)
results
```
